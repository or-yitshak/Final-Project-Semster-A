### Question 1.a.
#### Approximately $\frac{1}{125}$ of the births are unidentical twins and $\frac{1}{300}$ are identical twins. Elvis had a twin that passed away in his birth. what is the probability of Elvis to be an identical twins?

At first we will notice that:
$\frac{1}{125}=\frac{12}{1500}$ and $\frac{1}{300}=\frac{5}{1500}$
so the total probability of having a twin is: $\frac{5}{1500}+\frac{12}{1500}=\frac{17}{1500}$

using bayses theorem: $${P(H/E)}=\frac{P(E/H)*P(H)}{P(E)}$$ we will calculate the probability of Elvis having an identical twin.

in our qustion:
- E - probability of having a twin.
- H - probability of having an identical twin.
$${P(E/H)}=1$$

$${P(H)}=\frac{5}{1500}$$

$${P(E)}=\frac{17}{1500}$$

thereby:
$${P(H/E)}=\frac{1*\frac{5}{1500}}{\frac{17}{1500}} = \frac{5}{17}$$

### Question 4
#### Roi is playing a dice game with Yael. Roi will roll 2 six-sided dice, and if the sum of the dice is divisible by 3, he will win 6 dolars If the sum is not divisible by 3, he will lose 3 dolars. What is Roiâ€™s expected value of playing this game?

At first we will calculte the probability of the sum to be divisible by 3. notice that there are 4 options: 3, 6, 9, 12.
- the probability that the sum will be 3 is $\frac{2}{36}$
- the probability that the sum will be 6 is $\frac{5}{36}$
- the probability that the sum will be 9 is $\frac{4}{36}$
- the probability that the sum will be 12 is $\frac{1}{36}$
- hence the probability that the sum will be divisible by 3 is: 
$\frac{2}{36}+\frac{5}{36}+\frac{4}{36}+\frac{1}{36}=\frac{12}{36}=\frac{1}{3}$
- the probability that the sum will not be divisible by 3 is: $1-\frac{1}{3}=\frac{2}{3}$

in other words the probability of winning is $\frac{1}{3}$ and the probability of losing is $\frac{2}{3}$.

to calculate the expected value of this game we will use this formula: 
$$E(X)=\sum_{i=1}^k x_i*P(x_i)$$
- k - the number of different outcomes.
- x - the outcome
- P(x) - the outcome probability

in our case there are 2 outcomes:
- Winning 6 dollars with a probability of $\frac{1}{3}$
- Losing 3 dollars with a probability of $\frac{2}{3}$

thereby:
$$E(X)=6*\frac{1}{3}+(-3)*\frac{2}{3}=2-2=0$$

### Question 6
#### A division of a company has 200 employees, 40%, percent of which are male. Each month, the company randomly selects 8 of these employees to have lunch with the CEO. What are the mean and standard deviation of the number of males selected each month?

First of all notice that this question deals with a case of binomial distribution. lets summarize what we know:
- 40% out of 200 employees are males that means that there are 80 males and 120 females.
- 40% males means that each time we pick ememployee randomly the probability that he will be a male is 0.4 and the probability that she will be a female is 0.6

we will use the next formula to calculate the mean = $\mu$.
$$\mu=n*P(x)$$
- $n$ - The number of times you repeat on the experiment
- $P(x)$ - the probability of the event you are intersted in

in our case:
- n - the number of times we choose employee randomly, 8.
- P(x) - the probability he will be a male, 0.4.

therefore the mean is:
$$\mu=8*0.4=3.2$$

the formula to calculate standard deviation ($\sigma$) of binomial distribution is:
$$\sigma=\sqrt{n*P(x)*P(\bar x)}$$
- $n$ - The number of times you repeat on the experiment
- $P(x)$ - the probability of the event you are intersted in
- $P(\bar x)$ - the probability of the complementary event

in our case:
- n - the number of times we choose employee randomly, 8.
- P(x) - the probability he will be a male, 0.4.
- P(x) - the probability she will be a female, 0.6.

therefore the mean is:
$$\sigma=\sqrt{8*0.4*0.6}=1.3856$$
